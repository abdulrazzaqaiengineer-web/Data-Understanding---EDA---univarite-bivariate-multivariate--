Bivariate and Multivariate

```python
import seaborn as sns
titanic = pd.read_csv("/content/train.csv")
tips = sns.load_dataset("tips") # by default seaborn library ke andar ye data set majood ha
flight = sns.load_dataset("flights")
iris = sns.load_dataset("iris")
```

Scaterplot Numerical to Nmunerical

Bivariate

```python
sns.scatterplot(x=tips["total_bill"], y=tips["tip"])
```

Multivarite

```python
sns.scatterplot(x=tips["total_bill"], y=tips["tip"],hue = tips["sex"],style = tips["smoker"],size = tips["size"])
```

Barplot(Numerical - Catogorical)

```python
tips.head()
```

```python
sns.barplot(x=tips["sex"], y=tips["tip"])
```

```python
sns.barplot(x=tips["sex"], y=tips["tip"],hue = tips["time"])
```

Barplot(Numerical - Catogorical)

```python
sns.boxenplot(x = titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

Distplot Catogorical to numnerical

```python
sns.distplot(x=titanic[titanic['Survived']==0]["Age"],hist = False)
sns.distplot(x=titanic[titanic['Survived']==1]["Age"],hist = False)
```

Heatmap (Catogorical - Catogorical)

```python
pd.crosstab(titanic['Sex'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Sex'],titanic['Survived']))
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

Pairplot

```python
# pairplot automatically hamare dataset me se numerical columns ke scatterplot bna deta hah
sns.pairplot(iris,hue = "species")
```

Lineplot numerical to numerical

```python
sns.lineplot(x=flight['year'],y=flight['passengers'])
```

```python
# aik table bna dia hah jis me har saal ke har mahine ka data show ho raha ha
flight.pivot_table(values='passengers',index='month',columns='year')
```

```python
sns.heatmap(flight.pivot_table(values='passengers',index='month',columns='year'))
```
